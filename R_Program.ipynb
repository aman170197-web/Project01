````r

# -- input Section
# 1. Enter Student Name : Ravi
# 2. Enter Roll Number : RAVI009912
# 3. Enter Class : MCA
# 4. Enter JAVA Marks : 56
# 5. Enter C++ Marks : 89
# 6. Enter Python Marks : 12
# 7. Enter Ruby Marks : 56
# 8. Enter SQL Marks : 78

# -- Output Section
# Student Name : Ravi
# Roll Number : RAVI009912
# Class : MCA
# Enter JAVA Marks : 56/100
# C++ Marks : 89/100
# Python Marks : 12/100
# Ruby Marks : 56/100
# SQL Marks : 78/100
#    total = ?
#    Per % = ?%
#    grade = ?
````

````r
# -- Input Section
name <- readline("Enter Student Name: ")
roll <- readline("Enter Roll Number: ")
class <- readline("Enter Class: ")

java <- as.numeric(readline("Enter JAVA Marks: "))
cpp <- as.numeric(readline("Enter C++ Marks: "))
python <- as.numeric(readline("Enter Python Marks: "))
ruby <- as.numeric(readline("Enter Ruby Marks: "))
sql <- as.numeric(readline("Enter SQL Marks: "))

# -- Calculations
total <- java + cpp + python + ruby + sql
max_total <- 500
percentage <- (total / max_total) * 100

# -- Grade Calculation
if (percentage >= 90) {
  grade <- "A+"
} else if (percentage >= 80) {
  grade <- "A"
} else if (percentage >= 70) {
  grade <- "B"
} else if (percentage >= 60) {
  grade <- "C"
} else if (percentage >= 50) {
  grade <- "D"
} else {
  grade <- "F"
}

# -- Output Section
cat("\n--- Student Report ---\n")
cat("Student Name:", name, "\n")
cat("Roll Number:", roll, "\n")
cat("Class:", class, "\n\n")

cat("JAVA Marks:", java, "/100\n")
cat("C++ Marks:", cpp, "/100\n")
cat("Python Marks:", python, "/100\n")
cat("Ruby Marks:", ruby, "/100\n")
cat("SQL Marks:", sql, "/100\n\n")

cat("Total =", total, "/", max_total, "\n")
cat("Percentage =", round(percentage, 2), "%\n")
cat("Grade =", grade, "\n")
````

````r
# 1. Basic Salary Slip
# Write an R program to:

# Input Employee Name and Basic Salary

# Calculate:

# HRA = 20% of Basic Salary
# DA = 10% of Basic Salary
# Calculate Gross Salary

# Display a simple salary slip

employee_name <- readline("Enter Employee Name: ")
basic_salary <- as.numeric(readline("basic_salary: "))

# Define HRA and DA percentages
hra <- 20
da <- 10

hra_amount <- basic_salary * hra / 100
da_amount <- basic_salary * da / 100

gross_salary <- basic_salary + hra_amount + da_amount

# Display Salary Slip

cat("\n---Salary Slip---\n")
cat("Employee Name: ", employee_name, "\n")
cat("Basic Salary: ", basic_salary, "\n")
cat("HRA (", hra, "%) : ", hra_amount, "\n")
cat("DA (", da, "%) : ", da_amount, "\n")
cat("Gross Salary: ", gross_salary, "\n")
````

````r
# 2. Salary Slip with Fixed Allowances
# Create a program that:

# Takes Basic Salary as input

# Adds:

# HRA = 25%
# TA = ₹1500
# Prints Gross Salary

Employee_name <- readline("Enter Employee Name: ")
Basic_Salary <- as.numeric(readline("Basic Salary: "))

HRA <- 25
TA <- 1500

HRA_Amount <- Basic_Salary * HRA / 100

gross_salary <- Basic_Salary + HRA_Amount + TA

cat("\n---Salary Slip---\n")
cat("Employee Name: ", Employee_name, "\n")
cat("Basic Salary: ", Basic_Salary, "\n")
cat("HRA (", HRA, "%) : ", HRA_Amount, "\n")
cat("TA: ", TA, "\n")
cat("Gross Salary: ", gross_salary, "\n")
````

````r
# 3. Salary Slip with Deductions
# Write a program to:

# Input Basic Salary

# Calculate:

# HRA = 20%
# DA = 10%
# PF = 12% of Basic Salary
# Calculate Net Salary

Employee_name <- readline("Enter Employee Name: ")
Basic_Salary <- as.numeric(readline("Basic Salary: "))


HRA_percent <- 20
DA_percent <- 10

hra_amount <- Basic_Salary * HRA_percent / 100
da_amount <- Basic_Salary * DA_percent / 100
pf_amount <- Basic_Salary * 12 / 100 # PF is 12% of Basic Salary

Net_Salary <- Basic_Salary + hra_amount + da_amount - pf_amount

cat("\n----Salary Slip with Deductions---\n");
cat("Employee Name: ", Employee_name, "\n");
cat("Basic Salary: ", Basic_Salary, "\n");
cat("HRA (", HRA_percent, "%) : ", hra_amount, "\n");
cat("DA (", DA_percent, "%) : ", da_amount, "\n");
cat("PF (12%) : ", pf_amount, "\n");
cat("Net Salary: ", Net_Salary, "\n");
````

````r
# 4. Employee Salary Slip (Multiple Inputs)
# Take input of:

# Employee ID
# Employee Name
# Basic Salary
# Calculate:

# HRA = 30%
# DA = 15%
# Gross Salary
# Display all details in a formatted salary slip.

Employee_ID <- readline("Employee_ID : ")
Employee_name <- readline("Enter Employee Name: ")
Basic_Salary <- as.numeric(readline("Basic Salary: "))

HRA_percent <- 30
DA_percent <- 15

HRA_Amount <- Basic_Salary * HRA_percent /100
DA_Amount <- Basic_Salary * DA_percent /100
gross_salary <- Basic_Salary + HRA_Amount + DA_Amount

cat("\n--- Employee Salary Slip ---\n");
cat(sprintf("%-20s: %s\n", "Employee ID", Employee_ID));
cat(sprintf("%-20s: %s\n", "Employee Name", Employee_name));
cat(sprintf("%-20s: %.2f\n", "Basic Salary", Basic_Salary));
cat(sprintf("%-20s: %.2f\n", paste0("HRA (", HRA_percent, "%)"), HRA_Amount));
cat(sprintf("%-20s: %.2f\n", paste0("DA (", DA_percent, "%)"), DA_Amount));
cat(sprintf("%-20s: %.2f\n", "Gross Salary", gross_salary));
````

````r
# 5. Monthly Salary Slip
# Write a Python program that: # NOTE: This is an R program

# Takes Basic Salary

# Calculates:

# HRA = 20%
# Medical Allowance = ₹2000
# PF = 10%
# Prints Net Salary

Employee_name <- readline("Employee Name :")

Basic_Salary <- as.numeric(readline("Basic Salary :"))

cat("---Calculates---\n")

HRA_percent <- 20
Medical_Allowance_fixed <- 2000
PF_percent <- 10

HRA_Amount <- Basic_Salary * HRA_percent / 100

Medical_Allowance_Amount <- Medical_Allowance_fixed

PF_Amount <- Basic_Salary * PF_percent / 100

Net_Salary <- Basic_Salary + HRA_Amount + Medical_Allowance_Amount - PF_Amount

cat(sprintf("%-20s: %s\n","Employee Name ", Employee_name));
cat(sprintf("%-20s: %.2f\n","Basic Salary ", Basic_Salary));
cat(sprintf("%-20s: %.2f\n", paste0("HRA (", HRA_percent, "%)"), HRA_Amount));
cat(sprintf("%-20s: %.2f\n","Medical Allowance ", Medical_Allowance_Amount));
cat(sprintf("%-20s: %.2f\n", paste0("PF (", PF_percent, "%)"), PF_Amount));
cat(sprintf("%-20s: %.2f\n","Net Salary ",Net_Salary));
````

````r
# 8. Annual Salary Slip
# Write a program to:

# Input Monthly Basic Salary
# Calculate Annual Salary
# Display annual salary slip

Employee_name <- readline("Employee Name : ")
Basic_Salary <- as.numeric(readline("Basic Salary: " ))
Annual_Salary <- Basic_Salary * 12

cat("--Salary Slip--\n")
cat(sprintf("%20s: %s\n","Employee Name", Employee_name))
cat(sprintf("%20s: %.2f\n","Monthly Salary",Basic_Salary))
cat(sprintf("%20s: %.2f\n","Annual Salary",Annual_Salary))

````

````r
# 9. Salary Slip with Bonus
# Create a program that:

# Takes Basic Salary
# Adds Bonus = 5000
# Prints final salary

Employee_name <- readline("Employee Name : ")
Basic_Salary <- as.numeric(readline("Basic Salary : "))

Bonus <- 5000
Final_Salary <- Basic_Salary + Bonus

cat("\n---Salary Slip with Bonus---\n")
cat(sprintf("%-20s: %s\n","Employee Name ", Employee_name))
cat(sprintf("%-20s: %.2f\n","Basic Salary  ", Basic_Salary))
cat(sprintf("%-20s: %.2f\n","Bonus  ",Bonus))
cat(sprintf("%-20s: %.2f\n","Final Salary  ", Final_Salary))
````

````r
# 10. Salary Slip Mini Project (Beginner)
# Write a Python program that:

# Takes employee details
# Calculates allowances and deductions
# Displays a complete salary slip

Employee_name <- readline("Employee Name : ")
Basic_Salary <- as.numeric(readline("Basic Salary : "))

HRA <- 20
DA <- 30
PF <- 10
MEdical_allowance <- 5000


HRA_Amount <- Basic_Salary * HRA / 100
DA_Amount <- Basic_Salary * DA / 100
PF_Amount <- Basic_Salary * PF / 100

Final_Salary <- Basic_Salary + HRA_Amount + DA_Amount - PF_Amount

cat("\n---Salary Slip---\n")
cat(sprintf("%-20s: %s\n","Employee Name",Employee_name))
cat(sprintf("%-20s: %.2f\n","Basic Salary",Basic_Salary))
cat(sprintf("%-20s: %.2f\n",paste0("HRA (",HRA,"%)"),HRA_Amount))
cat(sprintf("%-20s: %.2f\n",paste0("DA(",DA,"%)"),DA_Amount))
cat(sprintf("%-20s: %.2f\n",paste0("PF(", PF,"%)"),PF_Amount))
cat(sprintf("%-20s: %.2f\n","Final Salary",Final_Salary))



````
